# Q: Which nutritional categories are closely correlated and not closely correlated and why?

```python
# Visualizes the relationships between several numerical attributes of each song using a correlation matrix heatmap

plt.figure(figsize=(12,12)) 
hm = sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt='.2f') 
plt.title('Starbuck Drinks Nutrional Data') 
plt.show()
```

A:  
The following are highly correlated at > 0.6:
- Calories & cholesterol(mg): 0.94, sugars(g): 0.91, trans fat(g): 0.64, total carbohydrates(g): 0.80
- Trans Fat(g) & Saturated Fat(g): 0.69, sodium(mg): 0.71
- Saturated Fat(g) & Sodium (mg): 0.92
- Total carbohydrates(g) & sugars(g): 0.77, cholesterol(mg): 0.77

The following are not highly correlated at < 0.0:
- Dietary Fibre(g) & Saturated Fat(g): -0.09, Sodium(mg): -0.06
- Vitamin C(% DV) & Trans Fat(g): -0.05, Saturated Fat(g): -0.04, Calcium(% DV): -0.05
- Iron(% DV) & Sodium(mg): -0.10, Saturated Fat(g): -0.07

Ingredients like trans & saturated fats, sugars, sodium , and cholesterol in large amounts usually go hand in hand in unhealthy strabucks drinks and thus it would make sense to see a strong correlation between there ingredients. Dietary fibre, saturated fat, sodium, iron, calcium, trans & saturated fat do not have a clear link between one another in food and beverages and thus would make sense if they were not found to be stringly correlated ingredients in the menu of beverages

# Q: Are the number calories present between signature espresso drinks and classic espresso drinks the same?



```python
# Tests to see if there is a normal distribution in the data to see if a two sample t-test is viable or a Mann-Whitley test if the data is not normally distributed

# DataFrame of the signature & classic espresso
sig_espresso = df[df['Beverage_category']=='Signature Espresso Drinks']
classic_espresso = df[df['Beverage_category']=='Classic Espresso Drinks']

# Uses the shapiro test to check for normal distribution
shapiro_sig_espresso = shapiro(sig_espresso['Calories'])
shapiro_classic_espresso = shapiro(classic_espresso['Calories'])

if shapiro_sig_espresso[1] < 0.05: 
    print('Reject the null hypothesis hence the data is not normally distributed')
else: print('Fails to reject null hypothesis hence the data is normally distributed')

if shapiro_classic_espresso[1] < 0.05: 
    print('Reject the null hypothesis hence the data is not normally distributed')
else: print('Fails to reject null hypothesis hence the data is normally distributed')
```

```python
# Tests to see whether null hypothesis can be rejected

ttest_results = ttest_ind(sig_espresso['Calories'], classic_espresso['Calories'])
if ttest_results[1] < 0.05: print('No, the number calories present between signature espresso drinks and classic espresso drinks is not the same')
else: print('Yes, the number calories present between signature espresso drinks and classic espresso drinks is the same')    
```
